```
import random

def guessing_game():
    secret_number = random.randint(1, 10)
    attempts = 0
    max_attempts = 3

    print("Bem-vindo ao Jogo de Adivinhação!")
    print(f"Estou pensando em um número entre 1 e 20. Você tem {max_attempts} tentativas.")

    while attempts < max_attempts:
        try:
            guess = int(input("Adivinhe o número: "))
            attempts += 1

            if guess < secret_number:
                print("Muito baixo!")
            elif guess > secret_number:
                print("Muito alto!")
            else:
                print(f"Parabéns! Você adivinhou o número {secret_number} em {attempts} tentativas.")
                break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")
    else:
        print(f"Suas {max_attempts} tentativas acabaram. O número secreto era {secret_number}.")

# Para iniciar o jogo, chame a função:
guessing_game()
```
